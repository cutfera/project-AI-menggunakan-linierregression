```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn
```

```python
dataset = pd.read_csv('DataGaji.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataku = pd.DataFrame(dataset)
plt.scatter(dataku.Tahun, dataku.Gaji)
plt.xlabel("Tahun")
plt.ylabel("Pendapatan")
plt.title("PROJECT AI CUT FERA INONGS SAFITRI")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = \
    train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color = 'green')
plt.title("PROJECT AI LinierRegression CUT FERA INONG SAFITRI")
plt.xlabel("Masa kerja")
plt.ylabel("Pendapatan")
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color = 'green')
plt.title("PROJECT AI LinierRegression Cut Fera Inong Safitri (TESTING TEST)")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```
